```python


import os
import cv2
def extract_features(img_path):
    # HoG 
    img = cv2.imread(img_path)
    img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    img_resized = cv2.resize(img_gray, (128, 128))
    hog = cv2.HOGDescriptor()
    features = hog.compute(img_resized)
    return features



start_path="C:/Users/dell/Downloads/Test Samples Classification/Test Samples Classification"

image_paths = []
labels = []

for root, dirs, files in os.walk(start_path):
    for dir_name in dirs:
        label = int(dir_name) - 1
        dir_path = os.path.join(root, dir_name)
        temp_images = os.listdir(dir_path)
        
        # Append image paths and labels
        image_paths.extend([os.path.join(dir_path, img) for img in temp_images])
        
        img_features = []
        for path in image_paths:
            img_features.append(extract_features(path))
            
        labels.extend([label] * len(temp_images))

```

```python
import joblib

model = joblib.load('svm_model.joblib')#deep learning model/
#model = load_model('C:/Users/dell/Desktop/intial-version-project-cv-using-cnn-with-deep/vision82e70a.h5')
predictions = model.predict(img_features)
print(predictions)
true_labels = labels
predicted_labels = np.argmax(predictions)
```
